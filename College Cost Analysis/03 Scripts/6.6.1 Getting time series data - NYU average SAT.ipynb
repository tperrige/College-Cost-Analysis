# 03. Re-formatting df-appended

```python
# Drop UNITID and INSTNM columns
df_appended.drop(columns = ['UNITID', 'INSTNM'], inplace = True)
```

```python
# Check output
df_appended
```

```python
# Switch order of columns
df_appended = df_appended[['DATE', 'SAT_AVG']]
```

```python
# Check output
df_appended
```

```python
# Change DATE to datetime format
df_appended['DATE'] = pd.to_datetime(df_appended['DATE'])
```

```python
# Sort DATE in chronological order
df_appended = df_appended.sort_values('DATE')
```

```python
# Set index to DATE
df_appended = df_appended.set_index('DATE')
```

```python
# Check output
df_appended
```

```python
# Export dataframe
df_appended.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'nyu_sat_time_series.pkl'))
```
